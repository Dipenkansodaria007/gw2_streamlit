```python
import requests
import pandas as pd
from datetime import datetime, timedelta

# Function to fetch data from the API
def fetch_data(endpoint):
    url = f"https://api.guildwars2.com/v2/{endpoint}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Failed to fetch data from {endpoint}. Status code: {response.status_code}")
        return None

# Fetch item pricing information
item_id = 19721
item_prices = fetch_data(f"commerce/prices/{item_id}")
if item_prices is None:
    exit()

# Display current selling price from item_prices
current_sell_price_api1 = item_prices['sells']['unit_price']
print("Current Sell Price (from API 1):", current_sell_price_api1)

# Fetch item historical pricing data
url = "https://api.datawars2.ie/gw2/v1/history?itemID=19721"
response = requests.get(url)
if response.status_code == 200:
    data = response.json()
    historical_data = pd.DataFrame(data)
    # Extract yesterday's date
    yesterday_date = (datetime.utcnow() - timedelta(days=1)).strftime('%Y-%m-%d')
    # Filter historical data for yesterday's date
    yesterday_data = historical_data[historical_data['date'].str.contains(yesterday_date)]
    if not yesterday_data.empty:
        yesterday_sell_price_avg = yesterday_data['sell_price_avg'].iloc[0]
        print("Yesterday's Sell Price Average (from API 2):", yesterday_sell_price_avg)
    else:
        print("No data found for yesterday's date.")
else:
    print(f"Failed to fetch historical data. Status code: {response.status_code}")
```

```python
import requests
import pandas as pd
from datetime import datetime, timedelta
import io

# Function to fetch data from the API
def fetch_data(endpoint):
    url = f"https://api.guildwars2.com/v2/{endpoint}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Failed to fetch data from {endpoint}. Status code: {response.status_code}")
        return None

# Fetch item pricing information from API 1
item_id = 19721
item_prices = fetch_data(f"commerce/prices/{item_id}")
if item_prices is None:
    exit()

# Display current selling price from API 1
current_sell_price_api1 = item_prices['sells']['unit_price']
print("Current Sell Price (from API 1):", current_sell_price_api1)

# Fetch item historical pricing data from API 2
url = "https://api.datawars2.ie/gw2/v2/history/hourly/csv?itemID=19721"
response = requests.get(url)
if response.status_code == 200:
    # Read CSV response into a DataFrame
    historical_data = pd.read_csv(io.StringIO(response.text))
    
    # Convert 'date' column to datetime
    historical_data['date'] = pd.to_datetime(historical_data['date'])
    
    # Calculate yesterday's date and time
    yesterday_date = (datetime.utcnow() - timedelta(days=1)).strftime('%Y-%m-%d')
    yesterday_hour = (datetime.utcnow() - timedelta(days=1)).hour
    
    # Filter historical data for yesterday's date and same time
    yesterday_same_time_data = historical_data[(historical_data['date'].dt.date == pd.to_datetime(yesterday_date).date()) & (historical_data['date'].dt.hour == yesterday_hour)]
    
    if not yesterday_same_time_data.empty:
        max_sell_price_same_time_yesterday = yesterday_same_time_data['sell_price_max'].max()
        print("Max Sell Price Same Time Yesterday (from API 2):", max_sell_price_same_time_yesterday)
    else:
        print("No data found for yesterday's date and same time.")
else:
    print(f"Failed to fetch historical data from API 2. Status code: {response.status_code}")
```
